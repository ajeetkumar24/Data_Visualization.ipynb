```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.
```

To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, you can use numpy, pandas, and seaborn. Here’s how to do it in Python

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 * 0.7 + np.random.normal(0, 0.1, n_samples)  # Correlated with x1
x3 = np.random.rand(n_samples)
x4 = x3 * 0.5 + np.random.normal(0, 0.1, n_samples)  # Correlated with x3

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population']*10,  # Bubble size proportional to Population
            alpha=0.6,
            edgecolors="w",
            linewidth=2)

# Annotate each bubble with the country name
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 fontsize=10, ha='center', va='center')

# Set labels and title
plt.title('Bubble Chart: Population vs. GDP by Country')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.xlim(0, 2200)
plt.ylim(0, 1000)
plt.grid(True)

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
import numpy as np

# Enable output in Jupyter notebook
output_notebook()

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a Bokeh figure
# Changed plot_width to width and plot_height to height
p = figure(title="Heatmap",
           x_axis_label='X',
           y_axis_label='Y',
           width=400,  # Changed from plot_width
           height=400) # Changed from plot_height

# Create a data source
source = ColumnDataSource(data={'x': np.repeat(x, 10),
                                 'y': np.tile(y, 10),
                                 'value': data_heatmap.flatten()})

# Create a color mapper
mapper = linear_cmap(field_name='value', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Add rectangles to the plot
# Refer to columns in the source for x and y
p.rect(x='x',  # Changed to refer to the 'x' column in source
        y='y',  # Changed to refer to the 'y' column in source
        width=0.1,
        height=0.1,
        source=source,
        fill_color=mapper,
        line_color='black')

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
